```python
# Read the file

import pandas as pd

fifa21 = pd.read_csv('fifa21_train.csv')

print(fifa21.shape)

pd.set_option('display.max_columns', None) # Setting the columns limit to none
fifa21.head(60)

fifa21.sort_values(by=['OVA'], axis = 0, ascending = False)
```

```python
#2 Lowercasing the column names

fifa21.columns = list(map(lambda x: x.lower(), fifa21.columns))
```

```python
#3 Checking for NaNs

nan = []

for col in fifa21.columns:
    if fifa21[col].isna().sum() != 0:
        print(col, fifa21[col].isna().sum())
# Some columns to take look into: club, joined, loan date end, the others might be dropped
```

```python
#4 Wrangling

import numpy as np

if 'loan date end' in fifa21.columns:
    fifa21 = fifa21.drop('loan date end', axis = 1) # dropping an entire column as its almost empty

fifa21['club'] = fifa21['club'].replace(np.nan,'Unknown') # Replacing the club with Unknown, we probably don't need this anyway
fifa21['joined'] = fifa21['joined'].replace(np.nan,1-1-2019) # Replacing the club with Unknown, we probably don't need this anyway

fifa21.shape # 11,701 rows at this point, 100 columns

fifa21 = fifa21.dropna() #Dropping other columns with NaN, they represent a small fraction of data
fifa21.shape # 11,222 rows at this point, 100 columns

removed = ((11701-11222)/11701)*100
round(removed,1) # by removing all NaNs, we removed 4.1% of data which is acceptable & easy

fifa21 = fifa21.reset_index(drop=True) #resetting the index so that loops with iterator can be used

# We are removing the +1, +2, +3 etc from ranking columns (ls, st, etc) - these additional pieces of information
# will be removed from model for now

for col in fifa21.loc[:,'ls':'gk']: 
    fifa21[col] = list(map(lambda x: x[:2], fifa21[col]))
    fifa21[col] = fifa21[col].str.replace('+', '.')
    fifa21[col] = pd.to_numeric(fifa21[col])

fifa21['gk'] = fifa21['gk'].astype(int)

for col in fifa21[['w/f','sm','ir']]: # Removing start signs from three columns
    fifa21[col] = list(map(lambda x: x[:1], fifa21[col]))
    fifa21[col] = pd.to_numeric(fifa21[col])
    
i = 0
for i in range(len(fifa21['hits'])): # Cleaning the 'hits' column
    if "K" in fifa21['hits'][i]:
        fifa21['hits'][i] = fifa21['hits'][i].replace('K','')
        fifa21['hits'][i] = pd.to_numeric(fifa21['hits'][i])
        fifa21['hits'][i] = fifa21['hits'][i]*1000
fifa21['hits'] = fifa21['hits'].astype(int)

if 'id' in fifa21.columns:
    fifa21 = fifa21.drop('id', axis = 1) # dropping the ID column, not needed and not a true numerical
    
i = 0
for col in fifa21[['a/w', 'd/w']]: # Encoding the categorical columns to be used
    for i in range(len(fifa21[col])):  
        if fifa21[col][i] == "Low":
            fifa21[col][i] = 0
        elif fifa21[col][i] == "Medium":
            fifa21[col][i] = 1
        elif fifa21[col][i] == "High":
            fifa21[col][i] = 2
```

```python
# Defining a function for the wrangling

def fifa21_clean(df):
    import numpy as np
    
    # Make all columns lowercase
    df.columns = list(map(lambda x: x.lower(), df.columns))
    
    #Remove 'loan end date' column as it is almost empty

    if 'loan date end' in df.columns:
        df = df.drop('loan date end', axis = 1)

    # Fill columns 'club' and 'joined' with dummy data; we don't want to remove them completely
    df['club'] = df['club'].replace(np.nan,'Unknown')
    df['joined'] = df['joined'].replace(np.nan,1-1-2019)
    
    # Drop rows with remaining NaNs - small portion of data
    df = df.dropna()
    
    # Reset index so the df can be used in for loops
    df = df.reset_index(drop=True)
    
    for col in df.loc[:,'ls':'gk']: # This removes the potential score for the players (+1, +2 etc) in the stats columns & turns into integer
        df[col] = list(map(lambda x: x[:2], df[col]))
        df[col] = df[col].str.replace('+', '.')
        df[col] = pd.to_numeric(df[col])
        
    df['gk'] = df['gk'].astype(int) # This changes the 'gk' column to int, which for some reason is converted to float first
    
    for col in df[['w/f','sm','ir']]: # Removing star signs from three columns
        df[col] = list(map(lambda x: x[:1], df[col]))
        df[col] = pd.to_numeric(df[col])     
        
    i = 0
    for i in range(len(df['hits'])): # Cleaning the 'hits' column
        if "K" in df['hits'][i]:
            df['hits'][i] = df['hits'][i].replace('K','')
            df['hits'][i] = pd.to_numeric(df['hits'][i])
            df['hits'][i] = df['hits'][i]*1000
    df['hits'] = df['hits'].astype(int)
    
    if 'id' in df.columns:
        df = df.drop('id', axis = 1) # dropping the ID column, not needed and not a true numerical  
        
    i = 0
    for col in df[['a/w', 'd/w']]: # Encoding the categorical columns to be used
        for i in range(len(df[col])):  
            if df[col][i] == "Low":
                df[col][i] = 0
            elif df[col][i] == "Medium":
                df[col][i] = 1
            elif df[col][i] == "High":
                df[col][i] = 2
    global cleandf
    cleandf = df
    return cleandf
```

```python
df = pd.read_csv('fifa21_train.csv')
fifa21_clean(df)
```

```python
#5 Creating categorical and numerical columns

numerical = fifa21.select_dtypes(include='number')
categorical = fifa21.select_dtypes(include='object')
categorical
```

```python
#6 Exploring numerical

numerical.shape # 53 columns are numerical

# Let's see the correlations

import seaborn as sns
import matplotlib.pyplot as plt


cf = 0.4
columns = []
corr_matrix = numerical.corr()

for col in numerical.columns:
        if abs(corr_matrix.loc[col,'ova']) >= cf:
            columns.append(col)

numerical_trimmed = pd.DataFrame(numerical, columns=columns)
numerical_trimmed

corr_matrix2 = numerical_trimmed.corr()
sns.set(rc={'figure.figsize':(15,8)})
sns.heatmap(corr_matrix2, annot=True)
plt.show()
```

```python
# 6 Normalizing numericals with MinMaxScaler

y = fifa21[['ova']]
X = fifa21.drop('ova', axis = 1)
X_num = X.select_dtypes(include='number')
X_cat = X.select_dtypes(include='object')

from sklearn.preprocessing import MinMaxScaler

Scaler = MinMaxScaler().fit(X_num)
X_normalized = Scaler.transform(X_num)
X_normalized = pd.DataFrame(X_normalized,columns=X_num.columns)

X_num_final = X_normalized
```

```python
#7 Encoding chosen categoricals

# The 'a/w' (away wins) and 'd/w' (domestic wins) categoricals have been encoded in step #4. Using LabelEncoder might also
# be possible, but we'd need to figure out the order of the labels (to have Low = 0, Medium = 1, High = 2)

X_encoded_final = X_cat[['a/w', 'd/w']]
```

```python
#8 Creating the model

X = pd.concat([X_num_final, X_encoded_final], axis = 1)

# Creating test-train split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 1337)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#9 Fitting the model

from sklearn import linear_model

lm = linear_model.LinearRegression()
lm.fit(X_train,y_train) # fitting the model with train data
```

```python
#10 Model Results

#6.1 R2

from sklearn.metrics import r2_score

predictions_train = lm.predict(X_train)
r2_train = r2_score(y_train, predictions_train)
print("R2 Train:",round(r2_train,2)) # interestingly, the R2_train is lower than R2_test!

predictions_test = lm.predict(X_test)
r2_test = r2_score(y_test, predictions_test)
r2_test
print('R2 Test:',round(r2_test,2))

#6.2 MSE

from sklearn.metrics import mean_squared_error

MSE = mean_squared_error(y_test,predictions_test)
print('MSE:',round(MSE,1))

#6.3 RMSE

import numpy as np

RMSE = np.sqrt(mean_squared_error(y_test,predictions_test))
print('RMSE:',round(RMSE,1))

#6.4 MAE

from sklearn.metrics import mean_absolute_error

MAE = mean_absolute_error(y_test, predictions_test)
print('MAE:',round(MAE,1))
```
